```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
mv['genres']=mv['genres'].apply(convert)
mv['keywords']=mv['keywords'].apply(convert)
mv.head()
```

```python
def convert3(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            l.append(i['name'])
            counter +=1
        else:
            break
        return l
```

```python
mv['cast']=mv['cast'].apply(convert3)
```

```python
mv.head()
```

```python
def fetch_directoer(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
    return l
    
```

```python
mv['crew']=mv['crew'].apply(fetch_directoer)
```
